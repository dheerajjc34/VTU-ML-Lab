Write a program to implement k-Nearest Neighbour algorithm to classify the iris data set. Print both correct and wrong predictions. Java/Python ML library classes can be used for this problem.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print(iris)
```

```python
print(iris.DESCR)
```

```python
X = iris.data
Y = iris.target
print("X\n", X[:5])
print("Y\n", Y[:5])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
print(iris.data.shape)
print("\nLength of:")
print("X_train : ", len(X_train))
print("Y_train : ", len(Y_train))
print("X_test : ", len(X_test))
print("Y_test : ", len(Y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test)
print(Y_pred)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
print("Accuracy : ", metrics.accuracy_score(Y_pred, Y_test))
print("\n")
print(confusion_matrix(Y_test, Y_pred))
print("\n")
print(classification_report(Y_test, Y_pred))
```

```python
predicted = knn.predict(X_test)
pre_target = [iris.target_names[i] for i in predicted]
print("Pre Target : ", pre_target, "\n\n")
actual_target = [iris.target_names[i] for i in Y_test]
print("Actual Target = ", actual_target, "\n\n")
print("\t Predicted\t\tActual\t\t\t\tAnswer")
print("-------------------------------------------------------------------------")
for i in range(0, len(pre_target)):
    print(i, ":", "\t", pre_target[i], "\t\t", actual_target[i], "\t\t", end="\t")
    if(pre_target[i] == actual_target[i]):
        print("Yes")
    else:
        print("No")
```

```python
s1 = [X_test[2]]
s = knn.predict(s1)
print(s)
s2 = [Y_test[2]]
print(s2)
if s == s2 :
    print("Correct!")
else:
    print("Wrong!")
```
